```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Heart (1).csv')
print(df.head())
```

# Scatter Plot

```python
plt.figure(figsize = (8,5))
sns.scatterplot(x=df["Age"], y=df["Chol"], hue=df["AHD"], palette="coolwarm")
plt.title("Age vs Cholesterol Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Cholesterol Level")
plt.show()
```

# Bar Plot

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='AHD', data=df, palette='pastel')
plt.title('Heart Disease Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

# Box Plot


```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='AHD', y='Chol', data=df, palette='Set2')
plt.title('Cholesterol Levels by Heart Disease')
plt.xlabel('Heart Disease (AHD)')
plt.ylabel('Cholesterol')
plt.show()
```

# Pie Chart

```python
plt.figure(figsize=(6, 6))
df['AHD'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Pie Chart: Distribution of Heart Disease (AHD)')
plt.ylabel('')
plt.show()
```

# Line Chart

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='MaxHR', data=df, marker='o', color='b')
plt.title('Line Chart: MaxHR vs Age')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.show()
```
