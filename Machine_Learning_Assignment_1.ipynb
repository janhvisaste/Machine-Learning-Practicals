# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_data = pd.read_csv('/content/Heart (1).csv')
csv_data.head()
```

```python
xls_data = pd.read_excel('/content/Heart (1).xlsx')
xls_data.head()
```

```python
print("Shape of CSV Data : ",csv_data.shape)
print("Shape of XLS Data : ",xls_data.shape)
```

```python
print("Missing values in CSV Data : ",csv_data.isnull().sum())
print("Missing values in XLS Data : ",xls_data.isnull().sum())
```

```python
print("Zeros in CSV Data:\n", (csv_data == 0).sum())
print("Zeros in Excel Data:\n", (xls_data == 0).sum())
```

```python
print("Selecting first 5 rows of CSV Data:\n", csv_data.head())
print("Sorting CSV Data by first column:\n", csv_data.sort_values(by=csv_data.columns[0]))
```

```python
print("Description of CSV Data:\n", csv_data.describe())
print("Description of Excel Data:\n", xls_data.describe())
```

```python
print("Unique values count in CSV Data:\n", csv_data.nunique())
print("Unique values count in Excel Data:\n", xls_data.nunique())
```

```python
csv_data['RestECG'] = csv_data['RestECG'].astype('float64')
print("Updated data type of RestECG:", csv_data['RestECG'].dtype)
```
